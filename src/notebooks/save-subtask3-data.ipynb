```python
import xml.etree.ElementTree as ET
```

```python
tree = ET.parse("../../data/subtask3-heterographic-test.xml")
root = tree.getroot()
```

```python
with open("../../data/subtask3-test.tsv", 'w') as f:
    for child in root:
        f.write(child.attrib['id'] + "\t")
        f.write(" ".join(word.text for word in child))
        f.write("\t")
        for word in child:
            if word.attrib['senses'] == "2":
                f.write(word.tex)
```
