```python
import operator
import xml.etree.ElementTree as ET
```

```python
lines = []
for line in data:
    if line and line != "\n":
        line_split = line.split(" ")[:3]
        lines.append((int(line_split[0][4:]), " ".join(line_split[1:])))
line_sorted = sorted(lines, key=operator.itemgetter(0))
```

```python
tree = ET.parse("../../data/subtask3-heterographic-test.xml")
root = tree.getroot()
```

```python
og_data = {}
for child in root:
    for word in child:
        if word.attrib['senses'] == "2":
            og_data[child.attrib['id']] = word.attrib['id']
```

```python
og_data['het_2']
```

```python
with open("../../subtask3_results_ordered", 'w') as out:
    for l in line_sorted:
        out.write(og_data["het_" + str(l[0])] + " " + l[1] + "\n")
```
