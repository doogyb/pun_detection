```python
import gzip
import os
from glob import glob
from pprint import pprint
```

```python
PATH = "/home/doogy/Data/3grams/c3/lower/"
PATH2 = "/home/doogy/Data/3grams/c3/upper"
```

```python
result = [y for x in os.walk(PATH) for y in glob(os.path.join(x[0], '*.gz'))]
```

```python
result.extend([y for x in os.walk(PATH2) for y in glob(os.path.join(x[0], '*.gz'))])
```

```python
from collections import defaultdict
```

```python
frequencies = defaultdict(lambda: defaultdict(lambda: defaultdict(int)))
```

```python
len(result)
```

```python
for i, file in enumerate(result):
    print(file)
    with gzip.open(file, 'rb') as f:
        for line in f.readlines():
            w1, w2, w3, freq = line.split()
            frequencies[w1.lower()][w2.lower()][w3.lower()] = int(freq)
        f.close()
```
