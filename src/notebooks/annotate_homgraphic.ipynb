```python
import xml.etree.ElementTree as ET
tree = ET.parse("../../data/subtask3-homographic-test.xml")
root = tree.getroot()
```

```python
limit = 0
puns = []
for child in root:
    words = " ".join(w.text for w in child)
    pun_word = list(filter(lambda x: x.attrib['senses'] == "2", [w for w in child]))[0].text
    puns.append((words, pun_word, child.attrib['id']))
    
puns
    
```

```python
from nltk.corpus import wordnet as wn
limit = 50
already_tagged = 2
with open("../../data/ani_annotations.tsv", 'a') as out_file:
    for count, line in enumerate(puns):
        if count <= already_tagged:
            continue
        if count > limit:
            break
        print("\n\n\n\t\t\t\t", count, "\n\n\n")
        print(line[0])
        original_word = line[1]
        print(original_word)
        senses = wn.synsets(original_word)
        if len(senses) == 0:
            continue
        for i, s in enumerate(senses):
            print(i, s.definition())
        syn1 = senses[int(input())]
        syn2 = senses[int(input())]
        
        out_file.write(line[2] + "\t" + syn1.lemmas()[0].key() + "\t" + syn2.lemmas()[0].key() + "\n")

        
```
