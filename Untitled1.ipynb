```python
def switch_score(distance, frequency_difference):
    if frequency_difference == 0:
        return 0
    return frequency_difference / ((distance + 1) ** 2)
```

```python
tp, fp = 0, 0
for i, replacements in enumerate(ranked_replacements):
#     print(replacements)
    scores = {switch_score(r[2], r[3]): r[1] for r in replacements}
#     print(scores)
    sorted_scores = sorted(scores.items(), key=operator.itemgetter(0), reverse=True)
#     print(sorted_scores)
#     break
    if len(sorted_scores) > 0:
        selected_word = sorted_scores[0][1]
    else:
        continue
    if selected_word == frequencies[i]['target']:
        tp += 1
    else:
        fp += 1
        print(i, selected_word, frequencies[i]['target'])
        
#         pprint(sorted_scores)
        
print("Accuracy: {}".format(tp / len(ranked_replacements)))
```
